## Importing Pandas and Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Task 1: Load the Dataset

```python
df = pd.read_csv("SalesDataCSV.csv")
df.head()
```

### OVERALL DATASET DISPLAY

```python
df
```

## Task 2: Explore and Clean the DataSet

```python
# Convert "Date" to datetime and drop empty rows in "Product"
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
df = df.dropna(subset=["Product"])  # Remove rows where "Product" is missing
df.loc[3, "Price"] = 700 # Adding Price of index 3 (Smartphone)
df.loc[5, "Price"] = 50 # Adding Price of index 5 (Jacket)
# Fill missing "Total" values using Quantity * Price
df["Total"] = df["Total"].fillna(df["Quantity"] * df["Price"])

df
```

## Task 3: Data Analysis and Manipulation

```python
top_5_products = df.sort_values(by="Total", ascending=False).head(5)
top_5_products
```

## Task 4: Data Visualization

### Bar Chart ( Total Sales for each Product)

```python
# Group by Product and calculate total sales
product_sales = df.groupby("Product")["Total"].sum().sort_values(ascending=False)

# Plot bar chart
plt.figure(figsize=(12, 6))
product_sales.plot(kind="bar", color="skyblue", edgecolor="black")

# Customize chart
plt.title("Total Sales per Product", fontsize=14)
plt.xlabel("Product", fontsize=12)
plt.ylabel("Total Sales ($)", fontsize=12)
plt.xticks(rotation=45, ha="right")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

### Line Chart ( Monthly Sales)

```python
# Extract month-year for trend analysis
df["Month"] = df["Date"].dt.to_period("M")
monthly_sales = df.groupby("Month")["Total"].sum()

# Plot line chart
plt.figure(figsize=(10, 5))
monthly_sales.plot(marker="o", linestyle="-", color="red")

# Customize chart
plt.title("Monthly Sales Trend", fontsize=14)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Total Sales ($)", fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, linestyle="--", alpha=0.7)
plt.show()
```
